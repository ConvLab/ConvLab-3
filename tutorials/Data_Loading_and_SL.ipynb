````python
import json
from convlab.util import load_dataset, load_ontology, load_database, load_dst_data
````

## Load Unified Dataloader object

Here we illustrate how DST data is loaded from the unified format. In a similar way users can load data for any module using the `load_*_data` module.

````python
dataset_name = "multiwoz21"
dataset = load_dataset(dataset_name, split2ratio={"train": 0.01})
````

````python
dst_data = load_dst_data(dataset, "test", "user", context_window_size=1)
print(json.dumps(dst_data['test'][1], indent=2))
````

## Load Unified ontology and database

Here we illustrate how to load the ontology and or dataset from the unified format. Here we show a extract from the multiwoz ontology.

````python
ontology = load_ontology(dataset_name)
database = load_database(dataset_name)
print(json.dumps(ontology['domains']['attraction'], indent=2))
````

Here we query the loaded databse using the dialogue state.

````python
# Dialogue state
state = {"hotel":
             {"area": "east",
              "price range": "moderate"}}

# Datbase query
res = database.query("hotel", state, topk=1)
print(json.dumps(res, indent=2))
````
