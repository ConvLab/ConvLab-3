```python
import json
```

```python
multi_domain_data = json.load(open('data/sgd/group1/multi_domain.json'))
```

```python
cross_domain_turns = []
for dial in multi_domain_data:
    for turn_idx in range(0, len(dial['turns']),2):
        turn = dial['turns'][turn_idx]
        if turn['cross_domain']:
            cross_domain_turns.append([turn['utterance'], turn['cross_domain_slot_value']])
```
